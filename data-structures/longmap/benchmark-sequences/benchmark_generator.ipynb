```python
import random
import math
def createOpSeqRandom(n):
    update = "update"
    remove = "remove"
    ops = [update, remove]
    opSeq = []
    for i in range(n):
        opSeq.append(random.choice(ops))
    
    res = ""
    for o in opSeq:
        l = f"{o} {random.randint(0, 460000000)} {'' if o == remove else str(random.randint(0, 460000000)) }"
        res += l+ "\n"
    return res

def createOpSeqAddNThenRemoveTheSame(n, randomised_order=False):
    update = "update"
    remove = "remove"
    keySeq = []
    for i in range(n):
        keySeq.append(random.randint(0, 460000000))
    
    res = ""

    # Add all
    if randomised_order:
        random.shuffle(keySeq)
    for k in keySeq:
        l = f"{update} {k} {random.randint(0, 460000000)}"
        res += l+ "\n"

    # Remove all
    if randomised_order:
        random.shuffle(keySeq)
    for k in keySeq:
        l = f"{remove} {k}"
        res += l+ "\n"
    return res

def createOpSeqAddNThenRemoveTheSameThenAddAgain(n, randomised_order=False):
    update = "update"
    remove = "remove"
    keySeq = []
    for i in range(n):
        keySeq.append(random.randint(0, 460000000))
    
    res = ""

    # Add all
    if randomised_order:
        random.shuffle(keySeq)
    for k in keySeq:
        l = f"{update} {k} {random.randint(0, 460000000)}"
        res += l+ "\n"
    
    # Remove all
    if randomised_order:
        random.shuffle(keySeq)
    for k in keySeq:
        l = f"{remove} {k}"
        res += l+ "\n"
    
    # Add all
    if randomised_order:
        random.shuffle(keySeq)
    for k in keySeq:
        l = f"{update} {k} {random.randint(0, 460000000)}"
        res += l+ "\n"
    return res

def createOpSeqAddNThenRemoveHalfThenUpdateAllAgain(n, randomised_order=False):
    update = "update"
    remove = "remove"
    keySeq = []
    for i in range(n):
        keySeq.append(random.randint(0, 460000000))
    
    res = ""
    # Add all
    if randomised_order:
        random.shuffle(keySeq)
    for k in keySeq:
        l = f"{update} {k} {random.randint(0, 460000000)}"
        res += l+ "\n"

    # Remove half
    i = 0
    if randomised_order:
        random.shuffle(keySeq)
    for k in keySeq:
        if i == math.floor(n/2):
            break
        l = f"{remove} {k}"
        res += l+ "\n"
        i += 1

    # Update all
    if randomised_order:
        random.shuffle(keySeq)
    for k in keySeq:
        l = f"{update} {k} {random.randint(0, 460000000)}"
        res += l+ "\n"
    return res


def write_to_file(p, content):
    with open(p, "w") as f:
        f.write(content)
```

```python
file_path = f"add-2-to-21-then-remove-half-then-add-again-randomised.txt"
write_to_file(file_path, createOpSeqAddNThenRemoveTheSameThenAddAgain(int(math.pow(2, 21))))
```
